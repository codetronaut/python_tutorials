+ **Assigning Values to Variables**
  
  _use "=" sign to assign values_

```python
x = 100                       # x is integer
pi = 3.14                     # pi is float
empname = "python is great"   # empname is string
 
a = b = c = 100 # this statement assign 100 to c, b and a.
```

## List of Common Operations

```python
list1 = [2, 3, 4, 1, 32]
print( 2 in list1 )
print(33 not in list1)
print( len(list1)) # find the number of elements in the list
print( max(list1)) # find the largest element in the list
print( min(list1)) # find the smallest element in the list
print( sum(list1)) # sum of elements in the list
```

## List slicing

```python
print(list1[0:3])

print(list1[:3])

print(list1[2:4])

print(list1[-1])

print(list1[-3:-1])
```

## List Operators

```python
print([1]+[2])
print([1]*3)
```

## Commonly used list methods with return type

```python
list1 = [2, 3, 4, 1, 32, 4]
list1.append(19)
print(list1)

list1.count(4) # Return the count for number 4
print(list1)

list1.index(4) # Return the index of number 4
print(list1)

list1.insert(1, 25) # Insert 25 at position index 1
print(list1)

list1.pop(2)
print(list1)

list1.remove(32) # Remove number 32
print(list1)

list1.reverse() # Reverse the list
print(list1)

list1.sort() # Sort the list
print(list1)
```

## Casting Variables

+ int() - constructs an integer number from an integer literal, a float literal (by rounding down to the previous whole number), or a string literal (providing the string represents a whole number)

+ float() - constructs a float number from an integer literal, a float literal or a string literal (providing the string represents a float or an integer)

+ str() - constructs a string from a wide variety of data types, including strings, integer literals and float literals


```python
int('1')
float('1.1')
str(4)

int(1.1)
float(1)

int('1.1')
```

## Set
+ unordered and unindexed collection

```python
a = set([1,87,6,'sfdsf',4,4.1,5,4,4,4,4])
a = {1,87,6,'sfdsf',4,4.1,5,4,4,4,4}

a.add("age")

a.remove(5.0)

for i in a:
    print(i)
```
